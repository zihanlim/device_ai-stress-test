# Poll Results — ERP Workload Tier Selection
**Date:** 2026-05-07

1 respondent selected **Option 2**, 6 respondents selected **Option 3**, 2 respondents selected **Option 4**.


```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

options = [
    'Option 1\nZoom + PPT + 10 Tabs + Excel',
    'Option 2\nOption 1 + PDF Reader + Outlook',
    'Option 3\nOption 2 + VSCode',
    'Option 4\nOption 3 + Large Excel (Power Query)\n+ Jupyter / SQL / Python',
    'Option 5\nOption 4 + Docker + Virtual Machines\n(UTM / Parallels)',
    'Option 6\nOption 5 + Local LLMs / PyTorch\n/ TensorFlow / RL',
]
counts = [0, 1, 6, 2, 0, 0]

# Unique color per option, matching Gantt chart config colors
OPT_COLORS = ['#6c5ce7', "#ff2aa3", "#0098b6", "#73028a", '#2e7d32', '#d32f2f']
colors = OPT_COLORS

fig, ax = plt.subplots(figsize=(12, 5), facecolor='white')
y_pos = list(range(len(options)))

LABEL_W = 5
COUNT_X = 4

bars = ax.barh(y_pos, counts,
               left=LABEL_W,
               color=colors, edgecolor='white', linewidth=0.25, height=0.35)

for y, count in zip(y_pos, counts):
    label_color = '#222222' if count > 0 else '#999999'
    ax.text(COUNT_X, y,
            str(count) if count > 0 else '0',
            ha='left', va='center', fontsize=10, color=label_color)

ax.set_yticks(y_pos)
ax.set_yticklabels([])
for y, opt in zip(y_pos, options):
    ax.text(0.05, y, opt, ha='left', va='center', fontsize=9.5, color='#222222',
            linespacing=1.3)

ax.invert_yaxis()

ax.set_xlim(0, LABEL_W + 9)
ax.set_xticks([int(LABEL_W) + i for i in range(0, 10)])
ax.set_xticklabels([str(i) for i in range(0, 10)])
ax.set_xlabel('Number of Respondents', fontsize=11)
ax.set_title('ERP Poll — Peak Workload Tier Distribution  (n=9 respondents)', fontsize=13, pad=12)
ax.set_facecolor('white')
ax.tick_params(length=0)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#999999')
ax.grid(axis='x', color='#f0f0f0', linewidth=0.8, zorder=0)
ax.set_axisbelow(True)

plt.tight_layout()
plt.subplots_adjust(left=0.07)
plt.savefig('poll_results.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print('Saved -> poll_results.png')
```
